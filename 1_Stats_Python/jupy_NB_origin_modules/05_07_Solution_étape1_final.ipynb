```python
def get(s, flag):
    res = re.split('\(|\,|\)',s)
    if flag == 0:
        res = res[1]
    else :
        res= res[2]   
    return float(res)  
```

```python
maDataFrame['X'] = maDataFrame['geometry'].apply(get, args=(0,))
```

```python
maDataFrame['Y'] = maDataFrame['geometry'].apply(get, args=(1,))
```
