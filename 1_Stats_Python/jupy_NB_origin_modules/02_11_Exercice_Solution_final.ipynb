```python
maDataFrame = maDataFrame.melt(['ID', 'nom', 'departement', 'ville', 'CP', 'type','geometry'], ['NO2_2012',
       'NO2_2013', 'NO2_2014', 'NO2_2015', 'NO2_2016', 'NO2_2017', 'PM10_2012',
       'PM10_2013', 'PM10_2014', 'PM10_2015', 'PM10_2016', 'PM10_2017',
       'PM25_2012', 'PM25_2013', 'PM25_2014', 'PM25_2015', 'PM25_2016',
       'PM25_2017'],'Polluant-Année','Valeur')
```

```python
maDataFrame
```

```python
maDataFrame['Polluant'] = maDataFrame['Polluant-Année'].apply(lambda s: s.split('_')[0])
```

```python
maDataFrame
```

```python
maDataFrame['Année'] = maDataFrame['Polluant-Année'].apply(lambda s: s.split('_')[1])
```

```python
maDataFrame
```

```python
def get(s, flag):
    res = re.split('\(|\,|\)',s)
    if flag == 0:
        res = res[1]
    else :
        res= res[2]   
    return float(res)  
```

```python
maDataFrame['X'] = maDataFrame['geometry'].apply(get, args=(0,))
```

```python
maDataFrame
```

```python
maDataFrame['Y'] = maDataFrame['geometry'].apply(get, args=(1,))
```
