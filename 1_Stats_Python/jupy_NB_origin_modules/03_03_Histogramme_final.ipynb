```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as pp
%matplotlib inline
```

```python
data = pd.read_csv('../../Data/NO2.csv')
```

```python
data
```

```python
pp.figure(figsize=(30,15))
pp.scatter(data.X,data.Y,color='b')
pp.scatter(data.X.mean(),data.Y.mean(),color='r')
```

```python
data['Région']='NE'
 
```

```python
data.loc[ (data.X<=data.X.mean()) & (data.Y>=data.Y.mean()) , 'Région']= 'NO'
```

```python
data.loc[ (data.X>data.X.mean()) & (data.Y<data.Y.mean())  , 'Région']= 'SE'
```

```python
data.loc[ (data.X<=data.X.mean()) & (data.Y<data.Y.mean())  , 'Région']= 'SO'
```

```python
data
```

```python
def plotAlldata():     
    colors = data.Région.map({'NO': 'r', 'NE': 'b', 'SE': 'g', 'SO': 'k'})   
    data.plot.scatter('X','Y',
                       c=colors,s=2,linewidths=0.5,edgecolor='k',alpha=0.5, figsize=(12,9))
    
```

```python
plotAlldata()
```
