```python
# EXERCICE1 : code function AVG, VARIANCE, STD AND COMPARE RESULT WITH NUMPY FUNCTIONS
import math # use only pow and sqrt
import numpy as np

uneliste = []
for i in range (1,31):
    uneliste.append(i)
    
# MY FUNCTIONS
def myAvg(ilist):
    sumlist=0
    for i in ilist:
        sumlist=sumlist+i
    return (sumlist/len(ilist))
    
def myVar(ilist):
    moyval = myAvg(ilist)
    calc = 0
    for i in ilist:
        calc = calc+ math.pow(i,2)
    return round(calc/len(ilist) - math.pow(moyval,2),2)

def myStd(ilist):
    var = myVar(ilist)
    return round((math.sqrt(var)),2)

print('avg myvalue = {} ; numpyvalue = {}'.format(myAvg(uneliste),np.mean(uneliste)))
print('var myvalue = {} ; numpyvalue = {}'.format(myVar(uneliste),np.var(uneliste)))
print('std myvalue = {} ; numpyvalue = {}'.format(myStd(uneliste),np.std(uneliste)))
```

**Définition des variables :**

Les particules fines (PM10 et PM 2,5) sont sans doute celles dont nous entendons le plus parler. Elles sont constituées d'un mélange hétérogène lié à l'usure de la route et des pneus, des garnitures de frein et d'embrayage, des hydrocarbures issus des pots d'échappement (fer, plomb, carbone) ou encore des additifs des lubrifiants et de l'oxyde d'aluminium provenant des supports des catalyseurs.

Le dioxyde d’azote NOx est le plus abondant. Il provient essentiellement de la circulation automobile et surtout actuellement des véhicules diesel, car à l’inverse des véhicules à essence ils ne sont pas équipés de pots d’échappement catalytiques détruisant ce polluant.

Le dioxyde d’azote (NO2) se forme dans l’atmosphère à partir du monoxyde d’azote (NO) qui se dégage essentiellement lors de la combustion de combustibles fossiles, dans la circulation routière par exemple. Le dioxyde d’azote se transforme dans l’atmosphère en acide nitrique, qui retombe au sol et sur la végétation

```python
maDataFrame.drop(columns=['geometry'],inplace=True) # useless column
```

```python
# MELT : TRANSFORM COLUMNS INTO ADDITIONNAL ROWS WITH COLUMN NAME > VALUE
# Function will keep first columnlist as it is
# will transform second columnlist into two new column and duplicate first columnlist value as necessary 

# input [12530:25] >> [225540:8]

maDataFrameMelt = maDataFrame.melt(['ID', 'nom', 'departement', 'ville', 'CP', 'type'], ['NO2_2012',
       'NO2_2013', 'NO2_2014', 'NO2_2015', 'NO2_2016', 'NO2_2017', 'PM10_2012',
       'PM10_2013', 'PM10_2014', 'PM10_2015', 'PM10_2016', 'PM10_2017',
       'PM25_2012', 'PM25_2013', 'PM25_2014', 'PM25_2015', 'PM25_2016',
       'PM25_2017'],'Polluant-Année','valeur')
maDataFrameMelt
```

```python
# GET ALL ROW CORRESPONDING TO QUERRY
maDataFrameMelt.query('ID=="0750026W"')
```

```python
# FUNCTION APPLY
# split column "Polluant-Année" into 2 separate column

# CREATE NEW COLUMN WITH DYNAMIC CONTENT : LAMBDA FUNCTION > split content of cell 
# and keep each side into its own column
maDataFrameMelt['Polluant'] = maDataFrameMelt['Polluant-Année'].apply(lambda s: s.split('_')[0])
maDataFrameMelt['Année'] = maDataFrameMelt['Polluant-Année'].apply(lambda s: s.split('_')[1])
maDataFrameMelt = maDataFrameMelt.drop(columns=['Polluant-Année'])
maDataFrameMelt
```

```python
# REORGANIZE DATA FRAME 
maDataFrameMelt = maDataFrameMelt[['ID', 'nom', 'departement', 'ville', 'CP', 'type','Année', 
                                   'Polluant', 'valeur']]
```

```python
# SAVE NEW REORGANIZE DATAFRAME INTO FILE
maDataFrameMelt.to_csv('data/output/ecoles-creches-idf-melt.csv', index=False)
```
