```python
import pandas as pd

# charger les données
data = pd.read_csv('decathlon.txt', sep="\t")

# éliminer les colonnes que nous n'utiliserons pas (3 dernieres)
# 10colonnes de perf sur xlignes de nom
my_data = data.drop(['Points', 'Rank', 'Competition'], axis=1)

# transformer les données en array numpy
X = my_data.values

print(X.shape)
data.head()
```

```python
import matplotlib.pyplot as plt
# AJOUT PERSO pour meilleure comprehension des données input
# DESCRIPTION SIMPLE DES DONNées 
for (columnName, columnData) in data.iteritems():
    #subset = data[columnName]
    print("-"*20)
    print(columnName)
    print("moy:\n",columnData.mean()) 
    print("med:\n",columnData.median()) 
    print("mod:\n",columnData.mode()) 
    print("var:\n",columnData.var(ddof=0)) #variance empirique (ddof variable pour definir estimation corrigée)
    print("ect:\n",columnData.std(ddof=0)) # ecart-type empirique
    columnData.hist(bins=10)
    plt.show()
    data.boxplot(column=columnName, vert=False) # delimitation quartile (vert determine position verticale)
    plt.show()
```

```python
from sklearn import preprocessing
# standardisation (hypothese données centrées)

#cherche le mean et standard dev du dataset (fit)
std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)
print(X_scaled.shape)
print(X_scaled)
```

```python
# visu SIMPLE DES DONNées transformées
# Xscaled est de la forme [[10];X], si on peut visualiser par epreuve il faut renverser le tableau
XscaledViz = X_scaled.T

#Loop on rows
for row in XscaledViz:
    print("-"*20)
    #print(columnName)
    print("moy:\n",row.mean()) 
    print("ect:\n",row.std(ddof=0)) # ecart-type empirique
    plt.hist(row, bins=10)
    plt.show()
    
# MEME VISUALISATION que données brute, mais valeurs centrées => MEAN =0 et STD = 1
```

```python
from sklearn import decomposition
#Calculs de composantes principales (sur N composantes)
# PCA =Principal component analysis (PCA)
pca = decomposition.PCA(n_components=2)
pca.fit(X_scaled)

# % de description par composantes principales
print(pca.explained_variance_ratio_)
# % de description total
print(pca.explained_variance_ratio_.sum())
```

```python
# projeter X sur les composantes principales
X_projected = pca.transform(X_scaled)

# on a maintenant toujours 41 rang, mais sur 2 colonnes
print(X_projected.shape)

plt.figure(figsize=(15,15))
# afficher chaque observation
# colorer en utilisant la variable 'Rank' pour juger de la position sur graphe en fonction de perfomance  
plt.scatter(X_projected[:, 0], X_projected[:, 1],
            c=data.get('Rank'))

# accrochage des noms des athletes surleur position dans le graphe
for i, (x,y) in enumerate(zip(X_projected[:, 0], X_projected[:, 1])):
    plt.text(x,y, data.index[i])
                          
plt.xlim([-5.5, 5.5])
plt.ylim([-4, 4])
plt.colorbar()
plt.show()
# (meilleur résultats plutot sur la droite du graphe (grosse valeur de la premiere composante))
```

```python
pcs = pca.components_
print(pcs)

#Contribution de chaque variable à chacune des deux composantes.

for i, (x, y) in enumerate(zip(pcs[0, :], pcs[1, :])):
    # Afficher un segment de l'origine au point (x, y)
    plt.plot([0, x], [0, y], color='k')
    # Afficher le nom (data.columns[i]) de la performance
    plt.text(x, y, data.columns[i], fontsize='14')

plt.xlabel('PC1')
plt.ylabel('PC2')
# Afficher une ligne horizontale y=0
plt.plot([-0.7, 0.7], [0, 0], color='grey', ls='--')

# Afficher une ligne verticale x=0
plt.plot([0, 0], [-0.7, 0.7], color='grey', ls='--')

plt.xlim([-0.7, 0.7])
plt.ylim([-0.7, 0.7])

# Les variables qui ont une contribution négative à la première composante principale (X) correspondent
# aux disciplines pour lesquelles une bonne performance est représentée par un petit nombre (temps de course),
# inversement pour les variables ayant une contribution positive (distances). 
# Cette composante permet de séparer les athlètes qui ont de très bonnes performances de 
# ceux qui sont (relativement !) plus mauvais dans toutes les disciplines.

# La deuxième composante principale (Y) permet de séparer les athlètes plus forts (bonnes performances aux lancers)
# et moins en sauts (pole et jump) des autres.

# Ce graphique nous permet aussi d'observer que certaines variables sont très corrélées :
# Discus (performance au lancer de disque) et Shot put (performance au lancer de poids)
# 100m et 100m.Hurdle
```
